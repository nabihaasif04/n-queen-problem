```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np

def is_safe(board, row, col, n):

    for i in range(row):
        if board[i][col] == 1:
            return False

    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i, j = row, col
    while i >= 0 and j < n:
        if board[i][j] == 1:
            return False
        i -= 1
        j += 1

    return True

def solve_n_queens_backtracking(n):
    board = [[0]*n for _ in range(n)]

    def backtrack(row):
        if row == n:
            return True
        for col in range(n):
            if is_safe(board, row, col, n):
                board[row][col] = 1
                if backtrack(row+1):
                    return True
                board[row][col] = 0
        return False

    backtrack(0)
    return board

def num_conflicts(board):
    n = len(board)
    conflicts = 0
    for i in range(n):
        for j in range(i+1, n):
            if board[i] == board[j]:
                conflicts += 1
            elif abs(i-j) == abs(board[i]-board[j]):
                conflicts += 1
    return conflicts

def hill_climbing(n, max_iter=1000):
    current = [random.randint(0, n-1) for _ in range(n)]

    for _ in range(max_iter):
        if num_conflicts(current) == 0:
            return current

        neighbors = []
        for row in range(n):
            for col in range(n):
                if col != current[row]:
                    neighbor = list(current)
                    neighbor[row] = col
                    neighbors.append(neighbor)


        current = min(neighbors, key=num_conflicts)

    return None

def fitness(chromosome):
    conflicts = 0
    n = len(chromosome)
    for i in range(n):
        for j in range(i+1, n):
            if chromosome[i] == chromosome[j] or abs(i-j) == abs(chromosome[i]-chromosome[j]):
                conflicts += 1
    return -conflicts

def genetic_algorithm(n, pop_size=100, generations=1000, mutation_rate=0.1):
    population = [[random.randint(0, n-1) for _ in range(n)] for _ in range(pop_size)]

    for _ in range(generations):
        population = sorted(population, key=fitness, reverse=True)
        if fitness(population[0]) == 0:
            return population[0]

        parents = population[:pop_size//10]

        children = []
        while len(children) < pop_size:
            parent1, parent2 = random.sample(parents, 2)
            split = random.randint(1, n-1)
            child = parent1[:split] + parent2[split:]

            if random.random() < mutation_rate:
                child[random.randint(0, n-1)] = random.randint(0, n-1)

            children.append(child)

        population = children

    return None

def plot_solution(board, title):
    n = len(board)
    chessboard = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            chessboard[i][j] = (i + j) % 2

    plt.figure(figsize=(6, 6))
    plt.imshow(chessboard, cmap='binary')

    if isinstance(board[0], list):
        for i in range(n):
            for j in range(n):
                if board[i][j] == 1:
                    plt.text(j, i, '♛', fontsize=20, ha='center', va='center', color='red')
    else:
        for row, col in enumerate(board):
            plt.text(col, row, '♛', fontsize=20, ha='center', va='center', color='red')

    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.show()

def test_algorithm(algorithm, n, runs=10):
    successes = 0
    total_time = 0

    for _ in range(runs):
        start_time = time.time()
        solution = algorithm(n)
        total_time += time.time() - start_time
        if solution is not None:
            successes += 1

    success_rate = (successes / runs) * 100
    avg_time = total_time / runs

    print(f"Algorithm: {algorithm.__name__}")
    print(f"Success Rate: {success_rate}%")
    print(f"Average Time: {avg_time:.4f} seconds")
    print("-" * 40)

    return solution

if __name__ == "__main__":
    n = 4

    print("=== Testing CSP (Backtracking) ===")
    csp_solution = test_algorithm(solve_n_queens_backtracking, n)
    if csp_solution:
        plot_solution(csp_solution, "CSP (Backtracking) Solution")

    print("=== Testing Hill Climbing ===")
    hill_solution = test_algorithm(hill_climbing, n)
    if hill_solution:
        plot_solution(hill_solution, "Hill Climbing Solution")

    print("=== Testing Genetic Algorithm ===")
    genetic_solution = test_algorithm(genetic_algorithm, n)
    if genetic_solution:
        plot_solution(genetic_solution, "Genetic Algorithm Solution")
```
